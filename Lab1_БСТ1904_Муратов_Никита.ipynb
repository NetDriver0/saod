Задание №2
Написать генератор случайных матриц(многомерных), который принимает
опциональные параметры m, n, min_limit, max_limit, где m и n указывают размер
матрицы, а min_lim и max_lim - минимальное и максимальное значение для
генерируемого числа .

```python
%%time
import numpy as np
import time
import random
try:
    m = int(input("Введите количество строк m "))
except ValueError:
    m = 50
try:
    n = int(input("Введите число столбцов n "))
except ValueError:
    n = 50
try:
    min_lim = int(input("Введите минимальное число min_lim "))
except ValueError:
    min_lim = -250
try:
    max_lim = int(input("Введите максимальное число max_lim "))
except ValueError:
    max_lim = 1000 + 13
A = np.empty((m, n))
for i in range(m):
    for j in range(n):
        A[i, j] = random.randint(min_lim, max_lim)
print("Ваш массив: ")
print(A)
```

Сортировка выбором

```python
%%time
def selectionSort(A):
    for i in range(0, len(A) - 1):
        for j in range(i + 1, len(A)):
            if A[j] < A[i]:
                A[i], A[j] = A[j], A[i]

import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
selectionSort(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```

```python
%%time
def insertion_sort(A):
    for i in range(1, len(A)):
        temp = A[i]
        j = i - 1
        while (j >= 0 and temp < A[j]):
            A[j + 1] = A[j]
            j = j - 1
        A[j + 1] = temp

import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
insertion_sort(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```

```python
%%time
def bubble(A):
    for j in range(len(A) - 1, 0, -1):
        for i in range(0, j):
            if A[i] > A[i + 1]:
                A[i], A[i + 1] = A[i + 1], A[i]

import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
bubble(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```

```python
%%time
def shell(A):
    step = len(A) // 2
    while step:
        for i, el in enumerate(A):
            while i >= step and A[i - step] > el:
                A[i] = A[i - step]
                i -= step
            A[i] = el
        step = 1 if step == 2 else int(step * 5.0 / 11)
        
import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
shell(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```

```python
def heapsort(A):
    build_max_heap(A)
    for i in range(len(A) - 1, 0, -1):
        A[0], A[i] = A[i], A[0]
        max_heapify(A, index=0, size=i)
        
def parent(i):
    return (i - 1)//2
 
def left(i):
    return 2*i + 1
 
def right(i):
    return 2*i + 2
 
def build_max_heap(A):
    length = len(A)
    start = parent(length - 1)
    while start >= 0:
        max_heapify(A, index=start, size=length)
        start = start - 1
        
def max_heapify(A, index, size):
    l = left(index)
    r = right(index)
    if (l < size and A[l] > A[index]):
        largest = l
    else:
        largest = index
    if (r < size and A[r] > A[largest]):
        largest = r
    if (largest != index):
        A[largest], A[index] = A[index], A[largest]
        max_heapify(A, largest, size)

import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
        
heapsort(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)



```

```python
def TournamentSort(array):
    A= array.copy()
    for i in range(len(A)):
        tournamentSort(A[i])
    return A
def tournamentSort(A):
    tree = [None] * 2 * (len(A) + len(A) % 2)
    index = len(tree) - len(A) - len(A) % 2

    for i, v in enumerate(A):
        tree[index + i] = (i, v)

    for j in range(len(A)):
        n = len(A)
        index = len(tree) - len(A) - len(A) % 2
        while index > -1:
            n = (n + 1) // 2
            for i in range(n):
                i = max(index + i * 2, 1) 
                if tree[i] != None and tree[i + 1] != None:
                    if tree[i][1] < tree[i + 1][1]:
                        tree[i // 2] = tree[i]
                    else:
                        tree[i // 2] = tree[i + 1]
                else:
                    tree[i // 2] = tree[i] if tree[i] != None else tree[i + 1]
            index -= n

        index, x = tree[0]
        A[j] = x
        tree[len(tree) - len(A) - len(A) % 2 + index] = None

import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1
tournamentSort(B)
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```

```python
def quick(A, start, end):
    if end - start > 1:
        p = partition(A, start, end)
        quick(A, start, p)
        quick(A, p + 1, end)
 
 
def partition(A, start, end):
    pivot = A[start]
    i = start + 1
    j = end - 1
 
    while True:
        while (i <= j and A[i] <= pivot):
            i = i + 1
        while (i <= j and A[j] >= pivot):
            j = j - 1
 
        if i <= j:
            A[i], A[j] = A[j], A[i]
        else:
            A[start], A[j] = A[j], A[start]
            return j
 
 
import numpy as np
B = np.empty(m*n)
C = np.empty((m, n))
z = 0
for x in range(m):
    for y in range(n):
        B[z] = A[x, y]
        z = z + 1        
quick(B, 0, len(A))
z = 0
for x in range(m):
    for y in range(n):
        C[x, y] = B[z]
        z = z + 1
print(C)
```
